```python
## R1
# Full distribution
metrics_r1full = [x_r1[:,i] for i in range(18)]

# Filtered distribution
stability_mask_r1 = x_r1[:,12] > 0.9
balance_mask_r1 = x_r1[:,1] > 0.9
diversity_mask_r1 = x_r1[:,2] > 0.9
combined_mask_r1 = stability_mask_r1 & balance_mask_r1 & diversity_mask_r1
x_r1_filtered = x_r1[combined_mask_r1]
metrics_r1_filtered = [x_r1[combined_mask_r1][:,i] for i in range(18)]

## R2
# Full distribution
metrics_r2full = [x_r2[:,i] for i in range(18)]

# Filtered distribution
rsqrt_mask_r2 = x_r2[:,6] > 0.9
x_r2_filtered = x_r2[rsqrt_mask_r2]
metrics_r2_filtered = [x_r2[rsqrt_mask_r2][:,i] for i in range(18)]

## R3
# Full distribution
metrics_r3full = [x_r3[:,i] for i in range(18)]

stability_mask_r3 = x_r3[:,12] > 0.9
balance_mask_r3 = x_r3[:,2] > 0.9
diversity_mask_r3 = x_r3[:,3] > 0.9
rsqrt_mask_r3 = x_r3[:,6] > 0.9
combined_mask_r3 = stability_mask_r3 & balance_mask_r3 & diversity_mask_r3 & rsqrt_mask_r3
x_r3_filtered = x_r3[combined_mask_r3]
metrics_r3_filtered = [x_r3[combined_mask_r3][:,i] for i in range(18)]
```
